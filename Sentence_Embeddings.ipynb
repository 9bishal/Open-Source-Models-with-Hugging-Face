✅ util.cos_sim()
This function computes the cosine similarity between the two embeddings:

python

from sentence_transformers import util

cosine_score = util.cos_sim(embedding1, embedding2)
It returns a value between -1 and 1:

1 → exactly the same direction (very similar)

0 → orthogonal (not similar)

-1 → opposite direction

```python
from sentence_transformers import SentenceTransformer, util

# Initialize model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Input sentences
sentences1 = [
    "This cat sits outside",
    "A man is playing guitar",
    "The movies are awesome"
]

sentences2 = [
    "The dog plays in the garden",
    "A woman watches TV",
    "The new movie is so great"
]

# Encode sentences to get embeddings
embedding1 = model.encode(sentences1, convert_to_tensor=True)
embedding2 = model.encode(sentences2, convert_to_tensor=True)

# Compute cosine similarity between sentence pairs
cosine_scores = util.cos_sim(embedding1, embedding2)

# Print results
for i in range(len(sentences1)):
    print(f"\nSentence 1: {sentences1[i]}")
    print(f"Sentence 2: {sentences2[i]}")
    print(f"Cosine Similarity Score: {cosine_scores[i][i]:.4f}")
```

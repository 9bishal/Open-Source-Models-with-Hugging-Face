##Build the chatbot pipeline using 🤗 Transformers Library


```python
from transformers.utils import logging
logging.set_verbosity_error()
```

```python
from transformers import pipeline
```

```python
from transformers import pipeline

chatbot = pipeline("text2text-generation", model="facebook/blenderbot-400M-distill")
```

```python
user_message = """
What are some fun activities I can do in the winter?
"""
```

```python
from transformers import BlenderbotTokenizer, BlenderbotForConditionalGeneration

model_name = "facebook/blenderbot-400M-distill"
tokenizer = BlenderbotTokenizer.from_pretrained(model_name)
model = BlenderbotForConditionalGeneration.from_pretrained(model_name)

# Store chat history (simulate conversation)
chat_history = ""

def chat(user_input):
    global chat_history
    # Append user input to history
    chat_history += f"Human: {user_input}\n"

    # Generate response
    inputs = tokenizer(chat_history, return_tensors="pt")
    output_ids = model.generate(**inputs, max_length=100)
    bot_reply = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Append bot response to history
    chat_history += f"Bot: {bot_reply}\n"
    return bot_reply

# Try it
print(chat("Hello!"))
print(chat("What's your name?"))
```
